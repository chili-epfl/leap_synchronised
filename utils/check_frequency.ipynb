```python
import numpy as np
np.set_printoptions(threshold=np.inf)
```

```python
with open('test2', 'r') as f:
    content = f.readlines()
content = [x.strip() for x in content]
content
```

```python
list_leap = []
i = 3
while(i < len(content)):
    list_leap.append(content[i])
    i = i + 5
# len(content)
list_leap
```

```python
list_num = []
for ele in list_leap:
    ele = ele.split(":")[-1]
    ele = ele.split()
    ele = ele[0]
    list_num.append(ele)
    print(ele)
```

```python
list_num_new = []
for ele in list_num:
    ele = int(ele)
    list_num_new.append(ele)
list_num_new
```

```python
list_sub = []
i = 1
while(i < len(list_num_new)):
    ele = list_num_new[i] - list_num_new[i - 1]
    list_sub.append(ele)
    i = i + 1
```

```python
list_array = np.array(list_sub)
list_array = list_array/1000000
list_array
```
