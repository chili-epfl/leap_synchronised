```python
import numpy as np
```

```python
with open('WacomData', 'r') as f:
    content = f.readlines()
content = [x.strip() for x in content]
content
```

```python
x = []
y = []

for each_line in content:
    data = each_line.split('|')
#     print(data)
    if(len(data) > 3):
        x.append(float(data[3]))
        y.append(float(data[4]))
```

```python
from matplotlib import pyplot

pyplot.scatter(x, y, marker = '.')
pyplot.show()
```
