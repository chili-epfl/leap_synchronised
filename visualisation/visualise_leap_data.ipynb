```python
def to_nparray(string_co):
    string_co = string_co[1:]
    string_co = string_co[:-1]
    string_co = string_co.replace(" ", "")
    string_arr = string_co.split(',')
    return np.array([float(string_arr[0]), float(string_arr[1]), float(string_arr[2])])
```
